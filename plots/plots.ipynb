```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['figure.dpi'] = 200
plt.rcParams['savefig.dpi'] = 200
#plt.rcParams["figure.figsize"] = (10,6)

from labellines import labelLines

plt.rc('axes', axisbelow=True)
```

P vs V

```python
L_exp = np.array([75,90,100,115,140])
L_exp = L_exp**2

P_exp = np.array([0.36,0.246,0.19,0.138,0.095])



V = np.linspace(60,150,100)
V = V**2

T = 180
kb = 0.1
N = 100
P = N*kb*T/V

plt.plot(V,P, linewidth=2, marker='o',color='r', markersize=0.001, linestyle='--',label="Te√≥rica", zorder=-1)
plt.scatter(L_exp,P_exp, c='k',label="Experimental", zorder=1)
plt.legend()
plt.grid()
plt.xlabel(r'Volumen ($L^2$)')
plt.ylabel(r'Presi√≥n ($P$)')
```

P vs T

```python
T_exp = np.array([159.95,180,199.964,219.7845,139.9205])

P_exp = np.array([0.1728,0.196,0.22,0.24,0.154])-0.015



T = np.linspace(120,240,100)

V = 100**2
kb = 0.1
N = 100
P = N*kb*T/V

plt.plot(T,P, linewidth=2, marker='o',color='r', markersize=0.001, linestyle='--',label="Te√≥rica", zorder=-1)
plt.scatter(T_exp,P_exp, c='k',label="Experimental", zorder=1)
plt.legend()
plt.grid(zorder=-1)
plt.xlabel(r'Temperatura ($T$)')
plt.ylabel(r'Presi√≥n ($P$)')
```

P vs N

```python
import pandas as pd
from scipy.stats import maxwell
```

```python
df_V = pd.DataFrame()
```

```python
path = 'velocities\steps100_particles100_V06_volume100.csv'

df = pd.read_csv(path, delimiter=';')

fig, ax = plt.subplots()

V = df['V']

sns.histplot(V, stat='density', binwidth=0.5, label='Experimental', color='#9dbbeb', edgecolor='#e9f2f1')
params = maxwell.fit(V, floc=0)

x = np.linspace(0, np.max(V)*1.3, 100)
y = maxwell.pdf(x, *params)

df_V['x100'] = x
df_V['y100'] = y
#beta = 1/(self.T*self.kb)
#mbv = np.exp(-V**2*beta/2)*2*np.pi*V**2/(2*np.pi)*beta
sns.lineplot(x=x, y=y, color='b', label='Ajuste', ax=ax)

#ax.set_title('Distribuci√≥n de velocidad')
ax.grid(True)
ax.set_xlabel('Velocidad ($V$)')
ax.set_ylabel('Densidad')
#ax.set_ylim([-0.3,5])
#ax.set_xlim(bottom=0)
ax.legend()

#### 6 colisioneees
```
